```python
#1. Write a simple program to compute the probability of a birthday match based on the number of people in the room. 
#Plot the results with the number of people on the  𝑥  axis and the corresponding probability on the  𝑦  axis. 
#At what point (number of people) does the probability cross  50% ?

res = []
p=1
for i in range(0,100):
    p = p*((365-i)/365)
    res.append(1-p)


indices = []
for i in range(1,101):
    indices.append(i)

df = pd.DataFrame(res, indices)
df.plot()
```

```python
#3.Now write a program to draw 1000 samples of size k, and store the sample means. Do this for  𝑘=1,5,10,20,30,50 . 
#You might want to use a numpy array for this. For example, np.empty([6,1000]) will create an empty 2D array with 6 rows 
#and 1000 columns. Each row can correspond to a different sample size and each column will be a sample mean from a sample of 
#that row's size."

#make a big list from which to sample
normalRandomSample = []
for i in range(100000):
    x = np.random.exponential(4,1) #changed dist to exponential from normal
    normalRandomSample.append(x)
normalRandomSample = np.array(normalRandomSample).tolist()

flat_list = []
for sublist in normalRandomSample:
    for item in sublist:
        flat_list.append(item)

#make the list of samples of size k=1
k1 = []
for i in range(1000):
    rs1 = random.sample(flat_list, 1)
    avg1 = mean(rs1)
    k1.append(avg1)
#k_1

#make the list of samples of size k=5
k5 = []
for i in range(1000):
    rs5 = random.sample(flat_list, 5)
    avg5 = sum(rs5)/len(rs5)
    k5.append(avg5)

#make the list of samples of size k=10
k10 = []
for i in range(1000):
    rs10 = random.sample(flat_list, 10)
    avg10 = sum(rs10)/len(rs10)
    k10.append(avg10)

#make the list of samples of size k=20
k20 = []
for i in range(1000):
    rs20 = random.sample(flat_list, 20)
    avg20 = sum(rs20)/len(rs20)
    k20.append(avg20)

#make the list of samples of size k=30
k30 = []
for i in range(1000):
    rs30 = random.sample(flat_list, 30)
    avg30 = sum(rs30)/len(rs30)
    k30.append(avg30)

#make the list of samples of size k=50
k50 = []
for i in range(1000):
    rs50 = random.sample(flat_list, 50)
    avg50 = sum(rs50)/len(rs50)
    k50.append(avg50)

#print the dataframe
df2 = pd.DataFrame([k1,k5,k10,k20,k30,k50])
df2.index = ['1','5','10','20','30','50']
df2
#np.empty([6,1000])
```

Now plot histograms of each row of your array (corresponding to a set of sample means of increasing sizes. Can you plot a grid of all 6 histograms in one figure, and add labels to all of them? 

```python
fib, axes = plt.subplots(nrows=6, ncols=1, figsize = (7,10))
axes = axes.flatten()

axes[0].hist(df2.iloc[0])
axes[1].hist(df2.iloc[1])
axes[2].hist(df2.iloc[2])
axes[3].hist(df2.iloc[3])
axes[4].hist(df2.iloc[4])
axes[5].hist(df2.iloc[5])

plt.tight_layout()

axes[0].set_title('k=1')
axes[1].set_title('k=5')
axes[2].set_title('k=10')
axes[3].set_title('k=20')
axes[4].set_title('k=30')
axes[5].set_title('k=50')

plt.show()
```

Find the mean and standard deviation for each of the distributions you're plotting.

```python
print("k=1: mean:", df2.iloc[0].mean(), '| stdev:', stat.stdev(df2.iloc[0]))
print("k=5: mean:", df2.iloc[1].mean(), '| stdev:', stat.stdev(df2.iloc[1]))
print("k=10: mean:", df2.iloc[2].mean(), '| stdev:', stat.stdev(df2.iloc[2]))
print("k=20: mean:", df2.iloc[3].mean(), '| stdev:', stat.stdev(df2.iloc[3]))
print("k=30: mean:", df2.iloc[4].mean(), '| stdev:', stat.stdev(df2.iloc[4]))
print("k=50: mean:", df2.iloc[5].mean(), '| stdev:', stat.stdev(df2.iloc[5]))
```

What result does your plot demonstrate?

It's clear that as the number of trials in a Exponential distribution goes up, the distribution becomes more and more normal centered around the lambda parameter. When we are drawing a small sample, the mean and stdev are approximately equal, which is what we'd expect from an exponential distribution. As the sample size grows however, the means stays around the same while the stdev shrinks to become more normal.
